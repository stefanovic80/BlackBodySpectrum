```python
h = 6.2607e-34#js
c = 3e8 #m/s
K_B = 1.38e-23 #JK^-1
LambdaN = np.arange(200, 1200, 1) 
Lambda = LambdaN*1e-9

fig = plt.figure(figsize =(14, 9))
ax = fig.add_subplot(111)
ax.set_xlim(LambdaN[0], LambdaN[-1])
ax.grid(True)
degrees = 50
plt.xticks( rotation = degrees)
plt.yticks( rotation = degrees)
ax.set_title("Black Body Spectrum", fontweight="bold")
ax.set_xlabel("$\lambda$ (nm)", fontweight="bold")
ax.set_ylabel("emissivity ($Watt/m^2nm$)", fontweight="bold")
plt.axvline(x = 380, linewidth = 2, color = 'rebeccapurple')
plt.axvline(x = 435, linewidth = 1, color = 'indigo')

plt.axvline(x = 500, linewidth = 1, color = 'blue')
plt.axvline(x = 520, linewidth = 1, color = 'green')

plt.axvline(x = 565, linewidth = 1, color = 'yellow')
plt.axvline(x = 590, linewidth = 1, color = 'orange')

plt.axvline(x = 625, linewidth = 1, color = 'red')
plt.axvline(x = 740, linewidth = 2, color = 'r')

def plotFunction():
    line, = ax.plot(LambdaN, emissivita, label='%.0f Kelvin' %T )
    point, = ax.plot(xM, M, 'xr', ms=10, mew = 2)
    plt.legend()

for T in range(6500, 2000, -250): #Kelvin
    A = 2*np.pi*h*c**2
    alpha = h*c/(K_B*T)
    emissivita = 1e-9*A*( np.exp( alpha/Lambda ) - 1 )**-1/Lambda**5

    M = max(emissivita)
    xM = np.where(emissivita == M)[0][0]
    xM = LambdaN[xM]
    plotFunction()

plt.savefig("/home/stefano/Desktop/SpettroCorpoNero")
```
